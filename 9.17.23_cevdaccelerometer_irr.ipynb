```r
#mvpa * irrr (adjusted) time periods = 1
summary(m1 <- glm.nb(pa_mv_day_1 ~ cevd_event + age_levels + male + alc_freq_filter + smok_stat + bmi_filter + full_screen_time_filter + employ_status_array_filter + comorb_sum+ gov_assistance_filter +dep_diag_yes, data = data))
(est <- cbind(Estimate = coef(m1), confint(m1)))
exp(est)
# Calculate the Bonferroni-corrected 95% confidence interval
alpha <- 0.05  # Original significance level
num_tests <- length(coef(m1))  # Number of coefficients being tested
# Apply Bonferroni correction
alpha_corrected <- alpha / num_tests
# Calculate the confidence interval using the corrected significance level
ci <- confint(m1, level = 1 - alpha_corrected)
# Print the corrected confidence interval
exp(ci)
summary_m1 <- summary(m1)
adjusted_p_values <- summary_m1$coefficients[, "Pr(>|z|)"]
# Print the Bonferroni-corrected confidence interval and adjusted p-values
print(exp(ci))
print(adjusted_p_values)
```
